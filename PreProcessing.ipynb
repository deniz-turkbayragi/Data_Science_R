```r
# Read the train.csv file from the provided link https://www.kaggle.com/c/home-data-for-ml-course/data?select=train.csv
#library("readxl")
#setwd('C:/Users/Deniz/Data/')
df <- read.csv(file.path("train.csv"))
```

```r
# Display the column names of the dataset
colnames(df)
```

```r
# Check number of NAs for each columns
colSums(is.na(df))
```

```r
# What is the highest NA number in column? (You need to write a code)
max(colSums(is.na(df)))
```

```r
# Drop Id column 
df$Id <- NULL
```

```r
# Display whole rows with uncertain number of BsmtCond (NA)
df[is.na(df$BsmtCond) == T,]
```

```r
# Replace NAs with mean or median for LotFrontage column and explain why did you choose? 
#mean and median values are almost the same, so we can use mean too.
df$LotFrontage[is.na(df$LotFrontage)] = mean(df$LotFrontage , na.rm = TRUE)
```

```r
# Divide the month(MoSold) to the quarters as first term, second term, third term and fouth term.  Assign the results new column "Term".
df$Term <- ifelse(df$MoSold <= 3, "first term", ifelse(df$MoSold <= 6, "second term", ifelse(df$MoSold <=9, "third term","fourth term"))) 
```

```r
# Get subset of dataframe. Display first 10 rows of Term and MoSold columns. 
head(subset(df, select = c("Term", "MoSold")), 10)
```

```r
# Add new column named Month to your dataframe 
# Assign corresponding month name of MoSold column. 
num = c()
for (i in 1:length(df$MoSold)){
    num = c(num, switch(as.character(df$MoSold[i]), "1" = "January", "2" = "February", "3" = "March" , "4" = "April" , "5" = "May" , "6" ="June" , "7" = "July" , "8" = "August" , "9" = "September" , "10" = "October" , "11" = "November" , "12" = "December" , "Unknown"))
}
df$Month <- num
```

```r
# Get subset of dataframe. Display last 10 rows of Month and MoSold columns. 
tail(subset(df, select = c("Month", "MoSold")), 10)
```
