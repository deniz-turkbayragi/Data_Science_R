```r
#Question-8
# Create a function named "dayToName".
dayToName=function(x){
  mod=x%%7   # taking modulus at every step
  if(mod==1){      
      day="Monday"
    }else if(mod==2) {
      day="Tuesday"
    }
  else if(mod==3){
    day="Wednesday"
  }
  else if(mod==4){
    day="Thursday"
  }
  else if(mod==5){
    day="Friday"
  }
  else if(mod==6){
    day="Saturday"
  }
  else
  {
    day="Sunday"
  }
    return (day)
}

df$dayName=apply(df[,3,drop=F],1,FUN = dayToName)
# apply function coerces data so we can prevent by using F  and 
#column  3 represents the column number in frame 
head(df,10)
```

```r
#library(memisc)

#Question-16
# Apply 1 to 7th degree non-linear regression to independent and dependent columns. 

nonlin_1 <- lm(dependent ~ poly(Independent, 1, raw = TRUE), data = df) 

nonlin_2 <- lm(dependent ~ poly(Independent, 2, raw = TRUE), data = df) 

nonlin_3 <- lm(dependent ~ poly(Independent, 3, raw = TRUE), data = df) 

nonlin_4 <- lm(dependent ~ poly(Independent, 4, raw = TRUE), data = df) 

nonlin_5 <- lm(dependent ~ poly(Independent, 5, raw = TRUE), data = df) 

nonlin_6 <- lm(dependent ~ poly(Independent, 6, raw = TRUE), data = df) 

nonlin_7 <- lm(dependent ~ poly(Independent, 7, raw = TRUE), data = df) 

# Compare your results. 

#Comparing the models with anova(). (In Anova table, the least mean square errors and the highest signifance should prefeable.)
anova(nonlin_1)
#Hence, as we go from linear model to quadratic model, we can see a little bit decrease in both SS and MS in residuals.
anova(nonlin_2)
#As we go from quadratic model to cubic model, we can see a signifance decrease in both SS and MS in residuals. And also regression model has higher significancy.
anova(nonlin_3)
#As we go from cubic model to quartic model, we do not see a great difference between values of SS and MS in residuals. They are almost same. In fact, we see a little decrease in significancy on the model.
anova(nonlin_4)
#As we go from quartic model to quintic model, we see a little bit decrease in both SS and MS in residuals. 
anova(nonlin_5)
#As we go from quintic model to hexic model, we do not see a significant change in the values. They're almost the same.
anova(nonlin_6)
#As we go from quintic model to septic model, we do not see a significant change in the values. They're almost the same.
anova(nonlin_7)
 
#In light of the above, after applying the cubic model(3rd polynomial) we could'nt see a significant change in both SS and MS in residuals. 


#Comparing summary outputs
#library(memisc)
mtable123 <- mtable("Model 1"=nonlin_1,"Model 2"=nonlin_2,"Model 3"=nonlin_3,"Model 4"=nonlin_4,"Model 5"=nonlin_5,"Model 6"=nonlin_6,"Model 7"=nonlin_7,summary.stats=c("Residual standard error","R-squared","F","P","n"))
mtable123
#Models that fit the data well should have higher R-squared values and lower residual standard error.
#We can see a significant decrease(more than half) in residual standard error, significat increase in R-square and F values while applying the cubic model(3rd).
#As we go to next model(4th), the results are almost the same with the previous model(3rd). In fact, 4th model has worser standard error.
#As we go to next models(5th,6th,7th), we can see a little bit(%1) decrease in residual standard error and little bit(%1) increase in R-square values. 
#Thus, cubic model can be used. We are not gaining significantly better results by adding higher orders.
```

```r
#Support your results with graphs.
#Plot of fitted vs residuals. (No clear pattern should show in the residual plot if the model is a good fit)
par(mfrow=c(2,2),cex=0.9)

a<-plot(fitted(nonlin_1),residuals(nonlin_1))
b<-plot(fitted(nonlin_2),residuals(nonlin_2))
#We can see similar patterns in both first and second term polynomials.
c<-plot(fitted(nonlin_3),residuals(nonlin_3))
d<-plot(fitted(nonlin_4),residuals(nonlin_4))
e<-plot(fitted(nonlin_5),residuals(nonlin_5))
f<-plot(fitted(nonlin_6),residuals(nonlin_6))
g<-plot(fitted(nonlin_7),residuals(nonlin_7))
#Other models have no clear pattern.
```

```r
#visualizing the models (linear,quadratic,cubic)
#library(ggplot2)
plot_lin<-geom_smooth(method='lm',se=FALSE,formula=y~x,colour="black")
plot_2nd_poly<-geom_smooth(method='lm',se=FALSE,formula=y~poly(x,2),colour="green")
plot_3rd_poly<-geom_smooth(method='lm',se=FALSE,formula=y~poly(x,3),colour="red")

ggplot(df,aes(Independent, dependent)) + geom_point(colour="blue") + plot_lin + plot_2nd_poly + plot_3rd_poly  

#As you can see, third-order polynomial has a better fit in the data.
```
